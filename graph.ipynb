```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

data = pd.read_csv('data.csv')
#select rows where country name is 'United States' or 'Euro area' or 'China'
data = data[(data['Country Name'] == 'United States') | (data['Country Name'] == 'European Union') | (data['Country Name'] == 'China') | (data['Country Name'] == 'Brazil')]
data.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data):
    # Drop column: '1960'
    data = data.drop(columns=['1960'])
    # Drop column: '2021'
    data = data.drop(columns=['2021'])
    # Drop column: '2022'
    data = data.drop(columns=['2022'])
    # Drop column: 'Unnamed: 67'
    data = data.drop(columns=['Unnamed: 67'])
    return data

data_clean = clean_data(data.copy())
data_clean.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data_clean):
    # Drop column: 'Indicator Code'
    data_clean = data_clean.drop(columns=['Indicator Code'])
    return data_clean

data_clean_1 = clean_data(data_clean.copy())
data_clean_1.head()
```

```python
brazil = data_clean_1[data_clean_1['Country Name'] == 'Brazil']
china = data_clean_1[data_clean_1['Country Name'] == 'China']
euro = data_clean_1[data_clean_1['Country Name'] == 'European Union']
usa = data_clean_1[data_clean_1['Country Name'] == 'United States']

#remove indexs of countries
china = china.reset_index(drop=True)
euro = euro.reset_index(drop=True)
usa = usa.reset_index(drop=True)
brazil = brazil.reset_index(drop=True)

brazil.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(brazil):
    # Drop column: 'Country Name'
    brazil = brazil.drop(columns=['Country Name'])
    # Drop column: 'Country Code'
    brazil = brazil.drop(columns=['Country Code'])
    # Drop column: 'Indicator Name'
    brazil = brazil.drop(columns=['Indicator Name'])
    return brazil

brazil_clean = clean_data(brazil.copy())
china_clean = clean_data(china.copy())
euro_clean = clean_data(euro.copy())
usa_clean = clean_data(usa.copy())
```

```python
usa_values = np.array(usa_clean)
china_values = np.array(china_clean)
euro_values = np.array(euro_clean)
brazil_values = np.array(brazil_clean)
usa_values = usa_values.flatten()
china_values = china_values.flatten()
euro_values = euro_values.flatten()
brazil_values = brazil_values.flatten()
xticks = [x for x in range(1961, 2021)]
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
fig.set_dpi(200)
fig.set_size_inches(6, 4)
ax.plot(xticks, usa_values, label='United States')
ax.plot(xticks, china_values, label='China')
ax.plot(xticks, euro_values, label='European Union')
ax.plot(xticks, brazil_values, label='Brazil')
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.xaxis.set_major_formatter('{x:.0f}')
ax.xaxis.set_minor_locator(MultipleLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(25))
plt.xlabel('Years')
plt.ylabel('fertilizer use (Mt/ha)')
plt.ylim(0,500)
plt.xlim(1960, 2020)
plt.legend(loc= 2, labelcolor='black')
plt.show()
```
